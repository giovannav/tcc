```python
import os
import cv2
import numpy as np

def save_average_color_images(source_directory, target_directory):
    # Create the target directory if it doesn't exist
    os.makedirs(target_directory, exist_ok=True)

    # List all image files in the source directory
    image_files = [f for f in os.listdir(source_directory) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    # Process each image
    for image_file in image_files:
        image_path = os.path.join(source_directory, image_file)

        # Load the image
        image = cv2.imread(image_path)

        if image is not None:
            # Calculate the average color
            mean_color = np.mean(image, axis=(0, 1))
            mean_color_int = mean_color.astype(int)

            # Create an image with the average color
            avg_color_image = np.full_like(image, mean_color_int, dtype=np.uint8)

            # Save the average color image to the target directory with the same name
            avg_color_image_path = os.path.join(target_directory, image_file)
            cv2.imwrite(avg_color_image_path, avg_color_image)

            print(f"Processed: {image_file}")
        else:
            print(f"Failed to load: {image_file}")

```

```python
save_average_color_images('camas_tiles_train/camendoim6sCenL1', 'camas_tiles_train_avg/camendoim6sCenL1')
save_average_color_images('camas_tiles_train/Frente+Umida', 'camas_tiles_train_avg/Frente+Umida')
save_average_color_images('camas_tiles_train/Fundo+Intermediaria', 'camas_tiles_train_avg/Fundo+Intermediaria')
save_average_color_images('camas_tiles_train/Meio+Seca', 'camas_tiles_train_avg/Meio+Seca')
```

```python
import cv2

def equalize_image(image):
    # Convert the image to grayscale if it's in color
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image

    # Equalize the histogram
    equalized = cv2.equalizeHist(gray)

    # If the original image was color, convert the equalized grayscale image back to color
    if len(image.shape) == 3:
        equalized = cv2.cvtColor(equalized, cv2.COLOR_GRAY2BGR)

    return equalized
```

```python
import cv2
import matplotlib.pyplot as plt

# Load an image
#image_path = 'camas_tiles_train/camendoim6sCenL1/camendoim6sCenL1 (1)_bottom_left.jpg'  # Replace with the actual path to your image
image_path = 'camas_tiles_train/Frente+Umida/Frente+Umida (3)_top_right.jpg'  # Replace with the actual path to your image

#image_path = 'path_to_image.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Equalize the image
equalized_image = equalize_image(image)

# Display the original and equalized images
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

```python
import cv2

def equalize_image_y(channel):
    # Equalize the histogram of the input channel
    equalized = cv2.equalizeHist(channel)
    return equalized

def equalize_image_rgb_to_ycrcb(image):
    # Convert the image to YCrCb color space
    ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
    
    # Split the channels
    y, cr, cb = cv2.split(ycrcb)
    
    # Equalize the Y channel
    equalized_y = equalize_image_y(y)
    
    # Merge the equalized Y channel back with the Cr and Cb channels
    equalized_ycrcb = cv2.merge((equalized_y, cr, cb))
    
    # Convert back to BGR color space
    equalized_bgr = cv2.cvtColor(equalized_ycrcb, cv2.COLOR_YCrCb2BGR)
    
    return equalized_bgr
```

```python
def equalize_image_y(channel):
    # Equalize the histogram of the input channel
    equalized = cv2.equalizeHist(channel)
    return equalized

def equalize_image_rgb_to_ycrcb(image):
    # Convert the image to YCrCb color space
    ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
    
    # Split the channels
    y, cr, cb = cv2.split(ycrcb)
    
    # Equalize the Y channel
    equalized_y = equalize_image_y(y)
    
    # Merge the equalized Y channel back with the Cr and Cb channels
    equalized_ycrcb = cv2.merge((equalized_y, cr, cb))
    
    # Convert back to BGR color space
    equalized_bgr = cv2.cvtColor(equalized_ycrcb, cv2.COLOR_YCrCb2BGR)
    
    return equalized_bgr

def convert_and_save_images(source_directory, target_directory):
    # Create the target directory if it doesn't exist
    os.makedirs(target_directory, exist_ok=True)

    # List all image files in the source directory
    image_files = [f for f in os.listdir(source_directory) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    # Process each image
    for image_file in image_files:
        image_path = os.path.join(source_directory, image_file)

        # Load the image
        image = cv2.imread(image_path)

        if image is not None:
            # Equalize and convert the image to YCrCb and back to RGB
            equalized_rgb_image = equalize_image_rgb_to_ycrcb(image)

            # Save the RGB image to the target directory with the same name
            rgb_image_path = os.path.join(target_directory, image_file)
            cv2.imwrite(rgb_image_path, equalized_rgb_image)

            print(f"Processed: {image_file}")
        else:
            print(f"Failed to load: {image_file}")

```

```python
convert_and_save_images('camas_tiles_train/camendoim6sCenL1', 'camas_tiles_train_YCRCB/camendoim6sCenL1')
convert_and_save_images('camas_tiles_train/Frente+Umida', 'camas_tiles_train_YCRCB/Frente+Umida')
convert_and_save_images('camas_tiles_train/Fundo+Intermediaria', 'camas_tiles_train_YCRCB/Fundo+Intermediaria')
convert_and_save_images('camas_tiles_train/Meio+Seca', 'camas_tiles_train_YCRCB/Meio+Seca')
```
